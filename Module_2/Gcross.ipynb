```python
import re
import random

# Set up the pattersn adn responses
patterns = {
    r'.*hello (.*)': 
    [
        "Hello, How can I aid you?",
    ],
    r'.*i feel (.*)': 
    [
        "Why do you feel {0}?",
        "What is making you feel {0}?"
    ],
    r'.*i am (.*)':
    [
        "Why are you {0}?",
        "Did anything make you feel {0}? Why was that the case?"
    ],
    r'.* (.*) sorry':
    [
        "It is ok, you don't need to apologize."
    ],
    r'.*yes (.*)':
    [
        "You sound confident in your answer.",
        "Can you explain that a little bit more?"
    ],
    r'.*yes':
    [
        "You sound confident in your answer.",
        "Can you explain that a little bit more?"
    ],
    r'.*no':
    [
        "Are you sure?",
        "Why not?"
    ],
    r'.*my name is (.*)': 
    [
        "Hello, {0}. How can I help you today?",
        "Nice to meet you, {0}. What brings you here?"
    ],
    r'.*you are (.*)': 
    [
        "Why do you think I am {0}?",
        "Does it matter to you what I am?",
    ],
    r'.*I (want|need|am looking for) (.*)': 
    [
        "Why do you {0} {0}?",
        "What would you do if you got {0}?",
        "Have you asked anyone else for {0}?",
    ],
    r'.*do you remember (.*)': 
    [
        "Of course I remember {0}. Tell me more about it.",
        "Why do you think I should remember {0}?",
    ],
    r'.*do you think (.*)': 
    [
        "I'm not sure, what do you think?",
        "Why don't you tell me what you think?",
    ],
    r'.*': 
    [
        "Please go on.",
        "Tell me more.",
        "I'm listening.",
    ]

}

# Generates random responses from the list
def generate_response(responses):
    return random.choice(responses)

# Matches the user input and generate a response based on what was received
def respond(input_text):
    for pattern, responses in patterns.items():
        match = re.match(pattern, input_text, re.IGNORECASE)
        if match:
            response = generate_response(responses)
            return response.format(*match.groups())
    return "I'm sorry, could you please rephrase that?"

# Main Loop
while True:
    user_input = input("User: ")
    if user_input.lower() == 'quit':
        break
    bot_response = respond(user_input)
    print("Bot:", bot_response)
```
